# 4. Outliers

```python
sns.boxplot(df['ROM'])
```

```python
Q1 = df['ROM'].quantile(0.25)
Q3 = df['ROM'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['ROM'] >= lower_bound) & (df['ROM'] <= upper_bound)]
```
